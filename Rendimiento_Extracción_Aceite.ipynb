# **Control de Rendimiento en Extracción de Aceite**

Este ejercicio es ideal para introducir float, int y la concatenación de strings.

**Planteamiento:**

Una planta procesadora de palma de aceite necesita calcular el **Índice de Rendimiento** de Extracción (IRE). El operario debe ingresar el peso total de los racimos de fruta fresca (RFF) procesados en kilogramos y la cantidad de aceite crudo obtenido en litros.

**Requerimientos:**

1. Solicitar el nombre del lote de producción (Tipo string).
2. Solicitar el peso de los racimos en kg (Tipo float).(Pr)
3. Solicitar los litros de aceite obtenidos (Tipo float).(lt)
4. Calcular el rendimiento sabiendo que la densidad aproximada del aceite es de 0.92 kg/L.(r)

$$Pa =lt * 0.92$$
$$r=(Pa/Pr) * 100$$

**Prueba del programa:**

Si se ingresan **1000 kg** de racimos y se obtienen **200 litros** de aceite, el programa debe informar que el rendimiento del lote es del **18.4%.**







````python
# ENTRADA:Solicitar el nombre del lote de producción (Tipo string).
#Solicitar el peso de los racimos en kg (Tipo float).(Pr)
#Solicitar los litros de aceite obtenidos (Tipo float).(lt)
nombre = input("Ingrese el nombre del lote: ")
pr = float(input("Ingrese el peso de los racimos en kg: "))
lt = float(input("Ingrese la cantidad de aceite crudo obtenido en litros: "))

#PROCESO: Calcular peso el aceite (Pa) y el rendimiento (r)
pa = lt * 0.92
r = (pa/pr) * 100

#SALIDA: Informar el rendimiento del lote
print("El rendimiento del lote", nombre, "es del", r, "%.")
````

A continuación, se explica cada línea del código del problema de Control de Rendimiento en Extracción de Aceite:

```python
nombre = input("Ingrese el nombre del lote: ")
```
*   **`nombre = input("Ingrese el nombre del lote: ")`**: Esta línea solicita al usuario que introduzca el nombre del lote de producción. La entrada del usuario se guarda como una cadena de texto en la variable `nombre`.

```python
pr = float(input("Ingrese el peso de los racimos en kg: "))
```
*   **`pr = float(input("Ingrese el peso de los racimos en kg: "))`**: Esta línea solicita al usuario el peso total de los racimos de fruta fresca (RFF) en kilogramos. La función `input()` lee la entrada como una cadena, y `float()` la convierte a un número de punto flotante (decimal) antes de almacenarla en la variable `pr`.

```python
lt = float(input("Ingrese la cantidad de aceite crudo obtenido en litros: "))
```
*   **`lt = float(input("Ingrese la cantidad de aceite crudo obtenido en litros: "))`**: Similar a la línea anterior, esta línea pide al usuario la cantidad de aceite crudo obtenido en litros. La entrada se convierte a un número de punto flotante usando `float()` y se guarda en la variable `lt`.

```python
pa = lt * 0.92
```
*   **`pa = lt * 0.92`**: Esta línea calcula el peso del aceite (`pa`) en kilogramos. Para ello, multiplica la cantidad de aceite en litros (`lt`) por la densidad aproximada del aceite (0.92 kg/L).

```python
r = (pa/pr) * 100
```
*   **`r = (pa/pr) * 100`**: Esta línea calcula el Índice de Rendimiento de Extracción (IRE), o simplemente el rendimiento (`r`). Se divide el peso del aceite obtenido (`pa`) por el peso de los racimos procesados (`pr`) y el resultado se multiplica por 100 para expresarlo como un porcentaje.

```python
print("El rendimiento del lote", nombre, "es del", r, "%.")
```
*   **`print("El rendimiento del lote", nombre, "es del", r, "%.")`**: Finalmente, esta línea muestra el resultado al usuario. Utiliza la función `print()` para concatenar varias cadenas de texto y las variables `nombre` y `r` para informar el rendimiento calculado del lote de producción.
